# Neural Network abstraction

## Connectionist model built from complex interactions of simple units

- **Basic Model**: Multi-layer perceptron (MLP). Dense connections of units across multiple layers:
    - **Neurons**: Basic units that receive input, process it, and pass on the output.
    - **Layers**: 
      - **Input Layer**: Where data enters.
      - **Hidden Layers**: Intermediate layers where computation happens.
      - **Output Layer**: Final layer that outputs results.
    - **Activation Function**: Determines the output of neurons; common functions include Sigmoid, Tanh, and ReLU.

## Advantages:
 - **Hardware-friendly**: Regular structures, well suited for optimization (tensor-based computations, parallelization).
 - **Scalability**: You can train **huge models** on **huge datasets** (trillions of parameters over TBs of data).
 - **They can manage unstructured data**: Usually you don't need **feature engineering**, they can extract them automatically from the raw data (images, audio, text...).

## Disadvantages:
 - **Data-hungry**: Their benefits are usually seen only on **extremely large** datasets.
 - **Black boxes**: Predictions cannot be explained or interpreted. Potential unexpected behaviors. No guarantees.
 - **Poor handling of discontinuities**: Smoothness is assumed in the feature space and output. Difficulties with discrete problems (rule-based classificiations, counting, etc.)

## 3: Time Series classification with Transformers

Time series analysis is pervasive on Artificial Intelligence and Machine Learning applications (energy, finances, medicine,...). It has some distinguishing features:
 - The time dimension can only be analyzed in **one direction** (future information cannot be exploited).
 - Problems are usually **highly unbalanced**.
 - **Continuous** analysis is a typical scenario.
 - Essential for building **personalized models**.

### NNs for Time Series in theory: Recurrent Neural Networks

The most mathematically sound model for dealing with temporal data are Recurrent Neural Networks (RNNS):
 - They include **backward connections** in the architecture to keep information from past observations.
 - They are **sequential** in nature. Past observations condition future predictions.
 - Very **difficult to train**. Specialized architectures such as [LSTM](https://www.mathworks.com/discovery/lstm.html) or [GRUs](https://en.wikipedia.org/wiki/Gated_recurrent_unit) try to mitigate practical issues.

<img src="https://www.ibm.com/content/dam/connectedassets-adobe-cms/worldwide-content/cdp/cf/ul/g/27/80/what-are-recurrent-neural-networks-combined.jpg" width="70%"/>

### NNs for Time Series in practice: Transformers

Sequential training has been a major drawback for scaling RNN models. That's the main reason for the appearance in 2017 of the most successful architecture of the last decade: The Transformer. Their basic advantages are:
 - **Parallel Processing:** Transformers efficiently process entire sequences at once, leading to faster computations and better performance with large datasets.
 - **Long-Range Dependencies:** Self-attention captures relationships between distant parts of a sequence more effectively than RNNs, which struggle with long-range dependencies.
 - **Scalability and Performance:** Transformers achieve state-of-the-art results in various tasks and scale well with increased data and model size, surpassing the capabilities of traditional RNNs.

This, together with the spectacular surge of Large Language Models (LLMs) for natural language processing applications (and the hype...), has led to a quick replacement of RNNs by Transformers in most domains.

### Building the model

We will implement the basic transfomer architecture as described in the foundational article ["Attention is all you need"](https://arxiv.org/abs/1706.03762).

<img src="https://upload.wikimedia.org/wikipedia/commons/3/34/Transformer%2C_full_architecture.png" style="margin:auto" width="60%"/>

Our model processes a tensor of shape `(batch size, sequence length, features)`,
where `sequence length` is the number of time steps and `features` is each input
timeseries.

We include residual connections, layer normalization, and dropout.
The resulting layer can be stacked multiple times.

The projection layers are implemented through `keras.layers.Conv1D`.

```python
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    # Attention and Normalization
    x = keras.layers.MultiHeadAttention(
        key_dim=head_size, num_heads=num_heads, dropout=dropout
    )(inputs, inputs)
    x = keras.layers.Dropout(dropout)(x)
    x = keras.layers.LayerNormalization(epsilon=1e-6)(x)
    res = x + inputs

    # Feed Forward Part
    x = keras.layers.Conv1D(filters=ff_dim, kernel_size=1, activation="relu")(res)
    x = keras.layers.Dropout(dropout)(x)
    x = keras.layers.Conv1D(filters=inputs.shape[-1], kernel_size=1)(x)
    x = keras.layers.LayerNormalization(epsilon=1e-6)(x)
    return x + res
```

The main part of our model is now complete. We can stack multiple of those
`transformer_encoder` blocks and we can also proceed to add the final
Multi-Layer Perceptron classification head. Apart from a stack of `Dense`
layers, we need to reduce the output tensor of the `TransformerEncoder` part of
our model down to a vector of features for each data point in the current
batch. A common way to achieve this is to use a pooling layer. For
this example, a `GlobalAveragePooling1D` layer is sufficient.

```python
def build_model(
    input_shape,
    head_size,
    num_heads,
    ff_dim,
    num_transformer_blocks,
    mlp_units,
    dropout=0,
    mlp_dropout=0,
):
    inputs = keras.Input(shape=input_shape)
    x = inputs
    for _ in range(num_transformer_blocks):
        x = transformer_encoder(x, head_size, num_heads, ff_dim, dropout)

    x = keras.layers.GlobalAveragePooling1D(data_format="channels_first")(x)
    for dim in mlp_units:
        x = keras.layers.Dense(dim, activation="relu")(x)
        x = keras.layers.Dropout(mlp_dropout)(x)
    outputs = keras.layers.Dense(1, activation="sigmoid")(x)
    return keras.Model(inputs, outputs)
```

### Train and evaluate

We do a standard training and evaluation:

```python
input_shape = x_train.shape[1:]

model = build_model(
    input_shape,
    head_size=256,
    num_heads=4,
    ff_dim=4,
    num_transformer_blocks=4,
    mlp_units=[128],
    mlp_dropout=0.4,
    dropout=0.5,
)

model.compile(
    loss=keras.losses.BinaryCrossentropy(from_logits=True),
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    metrics=["binary_accuracy"],
)
model.summary()
```

```python
callbacks = [keras.callbacks.EarlyStopping(patience=20, restore_best_weights=True)]

history = model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    epochs=150,
    batch_size=64,
    callbacks=callbacks,
)

model.evaluate(x_test, y_test, verbose=1)
```

It's interesting that, even if the model has half of the parameters of the one used for MNIST, it is computationally much harder to train! Still, we achieved some effective learning and a prediction accuracy over 80% on the test set.
